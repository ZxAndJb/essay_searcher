```python
import torch
def generate_casual_mask(seq_len):
    casual_mask = torch.ones((seq_len,seq_len)).tril()
    casual_mask = casual_mask.masked_fill(casual_mask == 0, float('-inf'))
    return casual_mask
    
```

```python
generate_casual_mask(20) == 1
```
